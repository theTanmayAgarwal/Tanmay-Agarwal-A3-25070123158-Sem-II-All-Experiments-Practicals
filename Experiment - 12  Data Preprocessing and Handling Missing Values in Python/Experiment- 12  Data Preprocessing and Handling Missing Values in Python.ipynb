```python
import pandas as pd
import numpy as np

#Step 1: Load the dataset
df = pd.read_csv("student_data.csv")

print("Original Data: ")
print(df)
```

```python
#Step 2: Replace "-" with NaN to mark missing values
df.replace("-", np.nan, inplace=True)
df
```

```python
from collections import defaultdict
#Convert Columns to numeric where needed

df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Marks"] = pd.to_numeric(df["Marks"], errors='coerce')

''' When converting data (like strings -> datetime),
if some values are invalid or cannot be converted,
instead of throwing and error, it will:
Replace those invalid values with NaT (Not a Time)'''

df
```

```python
# Step 3: Check Missing ValuesView

print("\nMissingValues:")
print(df.isnull().sum())
```

```python
# Fill missing Age with mean

df["Age"].fillna(df["Age"].mean(), inplace=True)
df
```

```python
# Fill missing Marks with median

df["Marks"].fillna(df["Marks"].median(),inplace=True)

df
```

```python
print(df.isnull().sum())
```

```python
# Step 5: Fix inconsistent text formatting

df["Department"] = df["Department"].str.upper()
df
```

```python
# Step 6: Convert Admission Date to standard Date Format

df["Admission date"] = pd.to_datetime(df["Admission date"], errors='coerce')
```

```python
# Step 7: Show Cleaned Data
print("\nCleaned Data:")
print(df)
```

```python
# Step 8: Save cleaned dataset
df.to_csv("cleaned_student_data.csv", index=False)

print("\nCleaned CSV File saved")
```

```python
# Working with missing values on DataSet-2

df=pd.read_csv('/content/Cars93.csv')
df
```

```python
#Inputing the Missing Value

#1. Luggage.Room
df ['Luggage.room'] = df['Luggage.room'].fillna(value=df['Luggage.room'].mean())
```
