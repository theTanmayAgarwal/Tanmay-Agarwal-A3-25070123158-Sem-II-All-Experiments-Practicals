### Problem 1a. Check whether a number is positive or negative

```python
# Get input from the user
number_a = float(input("Enter a number to check if it's positive or negative: "))

# Check if the number is positive or negative
if number_a >= 0:
    print(f"The number {number_a} is positive.")
else:
    print(f"The number {number_a} is negative.")
```

### Problem 1b. Check whether a number is positive, negative, or zero

```python
# Get input from the user
number_b = float(input("Enter a number to check if it's positive, negative, or zero: "))

# Check if the number is positive, negative, or zero
if number_b > 0:
    print(f"The number {number_b} is positive.")
elif number_b < 0:
    print(f"The number {number_b} is negative.")
else:
    print(f"The number {number_b} is zero.")
```

# Problem 2. Odd and Even Function

```python
number = int(input("Enter a number to check Odd or Even: "))


if number % 2 == 0:
        print(f"The number {number} is even.")
elif number % 2 != 0:
        print(f"The number {number} is odd.")
else:
    print(f"The number {number} is not a whole number. Odd/Even typically applies to integers.")
```

### Problem 5: Get input from user for 5 subjects and find average
#### Calculate grading using if-elif-else

```python
sub1 = float(input("Enter marks for subject 1: "))
sub2 = float(input("Enter marks for subject 2: "))
sub3 = float(input("Enter marks for subject 3: "))
sub4 = float(input("Enter marks for subject 4: "))
sub5 = float(input("Enter marks for subject 5: "))

# Calculate the average
average = (sub1 + sub2 + sub3 + sub4 + sub5) / 5

print(f"The average marks are: {average:.2f}")

# Calculate grading using if-elif-else
if average >= 90:
    grade = 'A+'
elif average >= 80:
    grade = 'A'
elif average >= 70:
    grade = 'B'
elif average >= 60:
    grade = 'C'
elif average >= 50:
    grade = 'D'
else:
    grade = 'F'

print(f"The grade is: {grade}")
```

### Problem 6: Check if a given year is a leap year

```python
# To get year number from user
year = int(input("Enter a year: "))


if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

### Problem 7: Check date validity and increment if valid

```python
# Input date in dd-mm-yyyy format
date = input("Enter date (dd-mm-yyyy): ")

# Split the date
parts = date.split("-")

# Check correct format
if len(parts) != 3:
    print("Invalid date")
    exit()

day = int(parts[0])
month = int(parts[1])
year = int(parts[2])

# Check leap year
leap = False
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    leap = True

# Find max days in the month
if month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12:
    max_days = 31
elif month == 4 or month == 6 or month == 9 or month == 11:
    max_days = 30
elif month == 2:
    if leap:
        max_days = 29
    else:
        max_days = 28
else:
    print("Invalid date")
    exit()

# Validate day
if day < 1 or day > max_days:
    print("Invalid date")
    exit()

# Increment date
day += 1

if day > max_days:
    day = 1
    month += 1

    if month > 12:
        month = 1
        year += 1

# Print incremented date
print("Incremented date:", f"{day:02d}-{month:02d}-{year}")
```

### Problem 9: Tax Calculation on the basis of Income.

```python
income = float(input("Enter your annual income (in INR): "))
tax = 0.0

if income <= 300000:
    tax = 0
elif income <= 600000:
    tax = (income - 300000) * 0.05
elif income <= 900000:
    tax = 15000 + (income - 600000) * 0.10  # 5% on 3L (15000) + 10% on next slab
elif income <= 1200000:
    tax = 15000 + 30000 + (income - 900000) * 0.15 # 5% on 3L (15000) + 10% on 3L (30000) + 15% on next slab
elif income <= 1500000:
    tax = 15000 + 30000 + 45000 + (income - 1200000) * 0.20 # 5% on 3L (15000) + 10% on 3L (30000) + 15% on 3L (45000) + 20% on next slab
else:
    tax = 15000 + 30000 + 45000 + 60000 + (income - 1500000) * 0.30 # Sum of taxes from previous slabs + 30% on remaining

print(f"Your Income is ₹{income:,.2f}, the calculated tax is ₹{tax:,.2f}.")
```
