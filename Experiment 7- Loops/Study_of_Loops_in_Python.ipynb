### Problem 1. Print 1 to 5 using loops
#### Print 'i' as long as 'i' is less than 6.

```python
i = 1
while (i<=5):
  print(i)
  i=i+1
```

###Problem 2. Take an input from a user for how many times he want to print

```python
n = int(input("Enter the number to be printed: "))

i= 1
while (i<=n):
  print(i, end=" ")
  i+=1
```

### Problem 3. Factorial of a number

```python
num = int(input("Enter an integer: "))

if num == 0:
    print("The factorial of 0 is 1.")
elif num < 0:
    print("Factorial does not exist for negative numbers.")
else:
    factorial = 1
    i = 1
    while i <= num:
        factorial *= i
        i += 1
    print(f"The factorial of {num} is {factorial}.")
```

### Problem 4. Fibonacci Series using while loop

```python
num = int(input("Enter a number for fibonacci iterations: "))

a = 0
b = 1
count = 0

while count < num:
        print(a, end=" ")
        c = a + b
        a = b
        b = c
        count += 1
```

### Problem 5. Reverse a number
#### Example:- 123 => 321

```python
num = int(input("Enter a number that you want to reverse: "))
rev=0

while num>0:
  digit = num % 10
  rev = rev * 10 + digit
  num//= 10

print("Reversed number=", rev)
```

### Problem 6. Check Palindrome number/string

```python
# Method No. 1

str = input("Enter to check palindrome: ")
i=0
j=len(str) - 1

while str[i] == str [j]:
  print(f"Yes {str} is a Palindrome")
  break
  i+=1
  j-=1

else:
  print(f"No {str} is not a Palindrome")
```

```python
# Method No. 2

st=input("Enter the string: ")
rev = st[::-1]
if st==rev:
  print(f"Yes, {st} is a Palindrome.")
else:
  print(f"No, {st} is not a Palindrome.")
```

### Problem 7. Count digits in a number

```python
#Method 1.

num = int(input("Enter a number: "))
count = 0
while num> 0:
  count+= 1
  num //= 10

print(f"The number of digits is {count}.")
```

```python
# Method 2.

num_str = int(input("Enter a number: "))
count_digits = len(str(num_str))
print(f"The number of digits in {num_str} is {count_digits}.")
```

### With the Break Statement we can stop the loop even if the while condition is true:

```python
# Exit the loop when i is 3:

i=1
while i<=6:
  print(i, end=" ")
  if i == 3:
    break

  i +=1
```

```python
# Password Validation

import re

while True:
    password = input("Enter your password: ")

    # Check length
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        continue

    # Check for at least one uppercase letter
    if not re.search("[A-Z]", password):
        print("Password must contain at least one uppercase letter.")
        continue

    # Check for at least one lowercase letter
    if not re.search("[a-z]", password):
        print("Password must contain at least one lowercase letter.")
        continue

    # Check for at least one digit
    if not re.search("[0-9]", password):
        print("Password must contain at least one digit.")
        continue

    # Check for at least one special character
    if not re.search("[@#$%^&+=]", password):
        print("Password must contain at least one special character (e.g., @, #, $, %).")
        continue

    print("Password is valid!")
    break # Exit loop if password is valid
```
