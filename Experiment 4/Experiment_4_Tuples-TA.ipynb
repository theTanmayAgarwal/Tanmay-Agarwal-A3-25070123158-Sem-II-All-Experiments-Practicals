```python
# Python Tuples
mytuple=("Tanmay","Adya","Kiara")
print(mytuple[1])
```

```python
mytuple=("Tanmay","Adya","Kiara")
mytuple[1]="khwaish"
print(mytuple[1])
```

```python
# Append Items in Tuples
mytuple=("Tanmay","Adya","Kiara")
mytuple.append("khwaish")
print(mytuple)
```

```python
# Add Tuple to a Tuple
mytuple=("Tanmay","Adya","Kiara")
mytuple2=("khwaish",)
mytuple+=mytuple2
print(mytuple)
```

```python
numbers = (3, 7, 1, 9, 4)
print(numbers[1::2])
```

```python
list1 = (20,30,40,50,60,70)

print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-1])
```

```python
list1 = (20,30,40,50,60,10,70)

print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-2])
print(list1[-2:-4:-2])
```

```python
exam_result = ("Mathematics", 85, "B")

# Unpack the tuple elements into separate variables
subject, marks, grade = exam_result

# Display the subject, marks, and grade
print(f"Subject: {subject}")
print(f"Marks Obtained: {marks}")
print(f"Grade: {grade}")

# Add if condition to check for 75 marks
if marks >= 75:
    print("Student scored 75 or more marks.")
else:
    print("Student scored less than 75 marks.")
```
