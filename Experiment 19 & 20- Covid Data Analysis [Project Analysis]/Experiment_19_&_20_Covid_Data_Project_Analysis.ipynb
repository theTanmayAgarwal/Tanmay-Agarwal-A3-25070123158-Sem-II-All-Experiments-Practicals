```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/covid_19_data.csv")
data.head()
```

```python
data.tail()
```

```python
data=data.drop(['SNo','Last Update'],axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
data['ObservationDate']= data['ObservationDate'].astype('datetime64[ns]')
data['Confirmed']= data['Confirmed'].astype('int64')
data['Deaths']= data['Deaths'].astype('int64')
data['Recovered']= data['Recovered'].astype('int64')
data.info()
```

```python
data.head()
```

```python
# Active Cases:
data['Active']= data['Confirmed']-data['Recovered'] - data['Deaths']
data.head(50)

data.iloc[50:100]
```

```python
# Last/Current Date:

data['ObservationDate'].max()
```

```python
# Fetch the latest date data
latest_data = data[data['ObservationDate'] == data['ObservationDate'].max()]
latest_data.head()
```

```python
latest_data.shape
```

```python
latest_data['Country/Region'].value_counts()
```

```python
latest_data['Country/Region'].nunique()
```

```python
latest_data['Country/Region'].unique()
```

```python
# Counts of cases for each country:

countries = latest_data.groupby("Country/Region")[["Confirmed","Deaths","Active"]].sum()
countries = countries.reset_index()
```

```python
countries
```

```python
#Cases in China

countries[countries['Country/Region']=="Mainland China"]
```

```python
#Cases in India

countries[countries['Country/Region']=="India"]
```

```python
#Cases in Spain

countries[countries['Country/Region']=="Spain"]
```

```python
#Cases in Germany

countries[countries['Country/Region']=="Germany"]
```

```python
# Plot on World Map

import plotly.express as px
import plotly.graph_objects as go
```

```python
world_map = px.choropleth(countries,locations = "Country/Region",locationmode="country names",
                          color="Confirmed",color_continuous_scale="YlGnBu",range_color=[1,10000000])
world_map
```

```python
#Top 20 Country Region

top = latest_data.groupby("Country/Region")[['Confirmed','Recovered']].sum().reset_index()
top = top.sort_values(['Confirmed'],ascending=False)
top.head()
```

```python
top_20 = top.head(20)
top_20
```

# Experiment 20- Spain Covid Cases Analysis

```python
spain = data[data['Country/Region']=='Spain'].copy()
spain
```

```python
# Find total number of States

spain['Province/State'].nunique()
```

```python
spain['Province/State'].unique()
```

```python
# Calculate the mode of 'Province/State'
mode_province_state = spain['Province/State'].mode()[0]
print(f"The mode of 'Province/State' is: {mode_province_state}")
```

```python
spain['Province/State'].unique()
```

```python
#Spain Latest Data:

spain_latest_data = spain[spain['ObservationDate'] == spain['ObservationDate'].max()]
spain_latest_data
```

```python
#Top 20 Province/Sate

top_state = spain_latest_data.groupby("Province/State")[['Confirmed','Recovered']].sum().reset_index()
top_state = top_state.sort_values(['Confirmed'],ascending=False)
top_state.head(20)
```

```python
top_state['Confirmed'].max()
```

```python
top_state[top_state['Confirmed']==top_state['Confirmed'].max()]['Province/State'].values[0]
```
