```python
''' Problem Statement 2: Three students choose elective subjects.
Find the common subjects chosen by all students using set operations.
Concept Used: Set Intersection (&)'''

# Define the subjects chosen by each student
student1_subjects = {"AMT", "PES", "TPCS", "EDA"}
student2_subjects = {"PES", "DCLD", "EDA", "English"}
student3_subjects = {"TPCS", "AMT", "EDA", "History"}

# Find the common subjects using set intersection
common_subjects = student1_subjects & student2_subjects & student3_subjects

print("Student 1's subjects:", student1_subjects)
print("Student 2's subjects:", student2_subjects)
print("Student 3's subjects:", student3_subjects)
print("Common subjects chosen by all students:", common_subjects)
```

```python
''' Problem Statement 3: One Set contains student in the EDC Club, another in the IEEE Club.
Find Students who are:
  1. In Both Clubs,
  2. In Only one Club
Concept Used: Intersection, Symmetric Difference'''

edc_club_members = {"Tanmay", "Drishti", "Jayvee", "Jahnvi", "Tannishtha"}
ieee_club_members = {"Drishti", "Aditya", "Jahnvi", "Rucha", "Tanmay"}

# Students in both clubs (Intersection)
students_in_both_clubs = edc_club_members.intersection(ieee_club_members)

# Students in only one club (Symmetric Difference)
students_in_only_one_club = edc_club_members.symmetric_difference(ieee_club_members)

print("EDC Club Members:", edc_club_members)
print("IEEE Club Members:", ieee_club_members)
print("\nStudents in Both Clubs:", students_in_both_clubs)
print("Students in Only One Club:", students_in_only_one_club)
```

```python
''' Problem Statement 4: Absent Students
Given a set of all students and a set of present students, find the absent students.
Concepts Used: Set Difference (-)'''

# Define the set of all students
all_students = {"Tanmay", "Drishti", "Jayvee", "Jahnvi", "Tannishtha", "Samiksha", "Rucha"}

# Define the set of present students
present_students = {"Tanmay", "Drishti", "Samiksha", "Jahnvi"}

# Find the absent students using set difference
absent_students = all_students - present_students

print("All Students:", all_students)
print("Present Students:", present_students)
print("Absent Students:", absent_students)
```

# Python Dictionaries
## Dictionaries are used to store data values in key value pairs
##
## A dictionary is a collection which is ordered*, changable and do not allow duplicates.

```python
# Creation of a Dictionary
mydict = {
    "brand": "Rolex",
    "model": "Submariner",
    "year": 2021
}

print(mydict)
```

```python
# Dictionary takes latest value given for key. It will not allow duplicates.
mydict = {
    "brand": "Rolex",
    "model": "Submariner",
    "year": 2021,
    "year": 2026
}

print(mydict)
```

```python
# Usage of len() Length function in a dictionary to find out members of the dictionary

mydict = {
    "brand": "Rolex",
    "model": "Submariner",
    "year": 2021,
    "year": 2026
}

print(len(mydict))
```

```python
# Type of a dictionary is <'dict'>

mydict = {
    "brand": "Rolex",
    "model": "Submariner",
    "year": 2021,
    "year": 2026
}

print(mydict)
print(type(mydict))
```

```python
# Changing Values in Dictionary:
watch = {
    "brand": "Rolex",
    "model": "Submariner",
    "year": 2021
}

print("Original dictionary:", watch)

watch["year"] = 2023
watch["model"] = "Daytona"

print("Updated dictionary:", watch)
```

```python
# Adding new key in dictionary

watch = {
    "brand": "Rolex",
    "model": "Submariner",
    "year": 2021
}
watch["Color"] = "Tiffany"

values_dict= watch.values()
print(values_dict)
print(watch)
```

```python
''' Problem Statement 2: A dictionary stores students names and marks.
Write a python program to search for a students marks using the name, If the student is found, display a suitable message. '''

student_marks = {
    "Tanmay": 85,
    "Drishti": 92,
    "Jayvee": 78,
    "Jahnvi": 95,
    "Rucha": 88
}

print("Student Marks:", student_marks)

student_name_to_search = input("Enter the name of the student to search for: ")

if student_name_to_search in student_marks:
    marks = student_marks[student_name_to_search]
    print(f"{student_name_to_search}'s marks: {marks}")
else:
    print(f"Student '{student_name_to_search}' not found in the records.")
```

```python
''' Problem Statement 3. A dictionary stores usernames and password. Write a python program to validate user login. '''

user_credentials = {
    "admin": "admin123",
    "user1": "pass123",
    "guest": "guestpass"
}

print("Available users for testing: admin, user1, guest")
username_input = input("Enter username: ")
password_input = input("Enter password: ")

if username_input in user_credentials:
    if user_credentials[username_input] == password_input:
        print(f"Login successful! Welcome, {username_input}.")
    else:
        print("Incorrect password. Please try again.")
else:
    print("Username not found. Please register or try again.")
```
