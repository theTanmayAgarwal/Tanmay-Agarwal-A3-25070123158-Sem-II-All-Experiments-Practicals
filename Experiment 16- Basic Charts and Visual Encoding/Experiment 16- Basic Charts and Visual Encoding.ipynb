```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Dataset 1 – Student Performance Data
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thru', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignment_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
df
```

## 📊 Matplotlib Visualizations — Dataset 1

```python
# 1. Line Chart (Trend) — Study Hours over Days
plt.plot(df['Days'], df['Study_Hours'], marker='o', color='green', linestyle='-.')
plt.title('Study Hours Over Days')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Advanced Line Chart — Study Hours and Marks Trend
plt.figure(figsize=(7, 4))
plt.plot(df['Days'], df['Study_Hours'], marker='*', label='Study Hours', color='green')
plt.plot(df['Days'], df['Marks'], marker='s', color='aqua', label='Marks')
plt.title('Study Hours and Marks Trend')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
plt.show()
```

🔹 Common Marker Types

| Marker | Meaning |
|--------|---------|
| `'o'`  | Circle ● |
| `'s'`  | Square ■ |
| `'^'`  | Triangle ▲ |
| `'x'`  | Cross ✖ |
| `'*'`  | Star ⭐ |

🔹 Common Color Options: `'blue'`, `'red'`, `'green'`, `'black'`, `'orange'`  
Short forms: `'b'` → blue, `'r'` → red, `'g'` → green, `'k'` → black

```python
# 2. Bar Chart — Marks Comparison by Day
plt.bar(df['Days'], df['Marks'], color='#FFD700', edgecolor='black', linewidth=2)
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
# Bar Chart — Study Hours vs Marks
plt.figure(figsize=(7, 4))
plt.bar(df['Study_Hours'], df['Marks'], color='pink', edgecolor='black')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# Bar Chart with Value Labels
bars = plt.bar(df['Days'], df['Marks'], color='#800080')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, y, str(y), ha='center', va='bottom')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show()
```

```python
# Grouped Bar Chart — Study Hours vs Marks
plt.figure(figsize=(7, 4))
x = np.arange(len(df['Days']))
width = 0.35
plt.bar(x - width / 2, df['Study_Hours'], width, label='Study Hours')
plt.bar(x + width / 2, df['Marks'], width, label='Marks')
plt.xticks(x, df['Days'])
plt.title('Study Hours vs Marks')
plt.xlabel('Days')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# 3. Histogram (Distribution) — Marks
hist_data = {
    'Marks': [50, 60, 55, 70, 65, 75, 80, 85, 90, 45, 68, 72, 77, 82, 88]
}
hist_df = pd.DataFrame(hist_data)
mean_value = np.mean(hist_df['Marks'])

plt.figure(figsize=(7, 4))
plt.hist(hist_df['Marks'], bins=3, color='cyan', edgecolor='black', alpha=0.5)
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label='Mean')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# 4. Scatter Plot — Study Hours vs Marks
plt.scatter(df['Study_Hours'], df['Marks'])
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# Conditional Scatter Plot — Result-wise
df['Result'] = ['Fail', 'Pass', 'Fail', 'Pass', 'Pass']
colors = ['red' if r == 'Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'], c=colors)
plt.title('Study Hours vs Marks (Result-wise)')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

# 🌟 Extra / Self Learning (Additional Visualizations)

This section demonstrates **advanced and creative visualizations** beyond the standard charts covered in Experiment 16. A richer dataset (Dataset 3) is created for better visual insights.

```python
# Dataset 3 — Extended Sales Data for Advanced Visualizations
np.random.seed(42)
days_list    = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] * 6
regions_list = (['North'] * 7 + ['South'] * 7 + ['East'] * 7 +
                ['West'] * 7 + ['North'] * 7 + ['South'] * 7)
categories_list = (['Food'] * 14 + ['Clothing'] * 14 + ['Electronics'] * 14)

n = 42
df3 = pd.DataFrame({
    'Day':       days_list[:n],
    'Region':    regions_list[:n],
    'Category':  categories_list[:n],
    'Sales':     np.random.randint(150, 500, n),
    'Profit':    np.random.randint(10,  100, n),
    'Customers': np.random.randint(20,  100, n),
    'Rating':    np.random.uniform(3.0, 5.0, n).round(1)
})
print(df3.head(10))
print(f"\nShape: {df3.shape}")
```

### 1. Pairplot — Relationships Between All Numerical Variables

```python
# Pairplot — shows pairwise relationships between all numeric columns
# Colored by Category to reveal group patterns
plt.figure()
sns.pairplot(df3[['Sales', 'Profit', 'Customers', 'Rating', 'Category']],
             hue='Category', diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle('Pairplot — Dataset 3', y=1.02, fontsize=14, fontweight='bold')
plt.show()
```

### 2. Correlation Heatmap — Identifying Variable Relationships

```python
# Heatmap of correlation matrix for numerical features
plt.figure(figsize=(7, 5))
corr = df3[['Sales', 'Profit', 'Customers', 'Rating']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, linecolor='white',
            cbar_kws={'shrink': 0.8})
plt.title('Correlation Heatmap — Dataset 3', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 3. Violin Plot — Distribution Shape + Summary Statistics

```python
# Violin plot shows the distribution of Sales for each Category
# Inner 'box' shows the quartiles; width shows density
plt.figure(figsize=(8, 5))
sns.violinplot(x='Category', y='Sales', data=df3,
               palette='Set2', inner='box', linewidth=1.5)
plt.title('Sales Distribution by Category (Violin Plot)', fontsize=13, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

### 4. Boxen Plot (Letter-Value Plot) — Detailed Distribution

```python
# Boxen plot is an enhanced box plot showing more quantiles
plt.figure(figsize=(8, 5))
sns.boxenplot(x='Region', y='Profit', data=df3,
              palette='Pastel1', linewidth=1.2)
plt.title('Profit Distribution by Region (Boxen Plot)', fontsize=13, fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

### 5. Strip Plot + Swarm Plot — All Individual Data Points

```python
# Strip Plot — individual points, slight jitter to avoid overlap
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.stripplot(ax=axes[0], x='Category', y='Customers', data=df3,
              palette='Set1', jitter=True, size=6, alpha=0.7)
axes[0].set_title('Strip Plot — Customers by Category', fontweight='bold')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Customers')

# Swarm Plot — non-overlapping points arranged to show distribution
sns.swarmplot(ax=axes[1], x='Category', y='Customers', data=df3,
              palette='Set1', size=5)
axes[1].set_title('Swarm Plot — Customers by Category', fontweight='bold')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Customers')

plt.suptitle('Strip vs Swarm Plot Comparison', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 6. FacetGrid — Multi-Panel Comparison Across Groups

```python
# FacetGrid creates a grid of subplots based on a categorical variable
# Each panel shows the Sales histogram for one Category
g = sns.FacetGrid(df3, col='Category', hue='Category',
                  palette='Set2', height=4, aspect=1)
g.map(sns.histplot, 'Sales', bins=6, edgecolor='white', linewidth=0.6)
g.set_axis_labels('Sales', 'Count')
g.set_titles(col_template='{col_name}', fontweight='bold')
g.add_legend()
g.figure.suptitle('Sales Distribution by Category (FacetGrid)',
                  fontsize=14, fontweight='bold', y=1.03)
plt.tight_layout()
plt.show()
```

### 7. KDE Plot — Kernel Density Estimation (Smooth Distribution)

```python
# KDE plot estimates the probability density of a variable
# Multiple KDEs on the same axes allow comparison across groups
plt.figure(figsize=(8, 5))
for cat, color in zip(['Food', 'Clothing', 'Electronics'], ['salmon', 'steelblue', 'seagreen']):
    subset = df3[df3['Category'] == cat]['Sales']
    sns.kdeplot(subset, label=cat, color=color, linewidth=2, fill=True, alpha=0.25)

plt.title('Sales KDE by Category (Dataset 3)', fontsize=13, fontweight='bold')
plt.xlabel('Sales')
plt.ylabel('Density')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

### 8. Countplot with Hue — Frequency of Categories

```python
# Countplot shows frequency counts, hue adds a second category dimension
plt.figure(figsize=(9, 5))
sns.countplot(x='Region', hue='Category', data=df3, palette='Set3', edgecolor='black')
plt.title('Count of Records: Region × Category (Dataset 3)',
          fontsize=13, fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Count')
plt.legend(title='Category', loc='upper right')
plt.tight_layout()
plt.show()
```

### 9. Barplot with Estimator — Aggregated Metric per Group

```python
# Barplot with estimator='mean' shows average Sales per Category
# Error bars represent 95% confidence intervals by default
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Mean Sales by Category
sns.barplot(ax=axes[0], x='Category', y='Sales', data=df3,
            estimator='mean', palette='Blues_d', edgecolor='black', capsize=0.1)
axes[0].set_title('Mean Sales by Category', fontweight='bold')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Mean Sales')

# Mean Profit by Region
sns.barplot(ax=axes[1], x='Region', y='Profit', data=df3,
            estimator='mean', palette='Greens_d', edgecolor='black', capsize=0.1)
axes[1].set_title('Mean Profit by Region', fontweight='bold')
axes[1].set_xlabel('Region')
axes[1].set_ylabel('Mean Profit')

plt.suptitle('Barplot with Estimator (Dataset 3)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 10. Lineplot with Markers & Style — Time-Series per Group

```python
# Seaborn lineplot with hue, markers and dashes for styled multi-group lines
# We use daily average Sales grouped by Category
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
df3['Day'] = pd.Categorical(df3['Day'], categories=day_order, ordered=True)
daily_cat = df3.groupby(['Day', 'Category'], as_index=False)['Sales'].mean()

plt.figure(figsize=(10, 5))
sns.lineplot(data=daily_cat, x='Day', y='Sales', hue='Category',
             style='Category', markers=True, dashes=True,
             palette='tab10', linewidth=2.5, markersize=8)
plt.title('Average Daily Sales by Category (Lineplot with Markers)',
          fontsize=13, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Average Sales')
plt.legend(title='Category')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### 11. Subplots — Multi-Chart Dashboard in a Single Figure

```python
# A 2×3 dashboard combining multiple chart types in one figure
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
fig.suptitle('Dataset 3 — Multi-Chart Dashboard', fontsize=16, fontweight='bold', y=1.01)

# Panel 1 — Box Plot: Sales by Category
sns.boxplot(ax=axes[0, 0], x='Category', y='Sales', data=df3, palette='Set2')
axes[0, 0].set_title('Sales by Category (Box Plot)')
axes[0, 0].set_xlabel('Category'); axes[0, 0].set_ylabel('Sales')

# Panel 2 — Bar Plot: Mean Customers by Region
region_avg = df3.groupby('Region')['Customers'].mean().reset_index()
axes[0, 1].bar(region_avg['Region'], region_avg['Customers'],
               color=['#FF6B6B','#4ECDC4','#45B7D1','#96CEB4'], edgecolor='black')
axes[0, 1].set_title('Mean Customers by Region (Bar Chart)')
axes[0, 1].set_xlabel('Region'); axes[0, 1].set_ylabel('Avg Customers')

# Panel 3 — Histogram: Profit Distribution
axes[0, 2].hist(df3['Profit'], bins=8, color='lightcoral', edgecolor='black', alpha=0.8)
axes[0, 2].axvline(df3['Profit'].mean(), color='darkred', linestyle='--', label='Mean')
axes[0, 2].set_title('Profit Distribution (Histogram)')
axes[0, 2].set_xlabel('Profit'); axes[0, 2].set_ylabel('Frequency')
axes[0, 2].legend()

# Panel 4 — Scatter Plot: Sales vs Profit colored by Region
region_palette = {'North':'red','South':'blue','East':'green','West':'orange'}
for region, group in df3.groupby('Region'):
    axes[1, 0].scatter(group['Sales'], group['Profit'],
                       c=region_palette[region], label=region, alpha=0.7, s=60)
axes[1, 0].set_title('Sales vs Profit by Region (Scatter)')
axes[1, 0].set_xlabel('Sales'); axes[1, 0].set_ylabel('Profit')
axes[1, 0].legend(title='Region', fontsize=8)

# Panel 5 — KDE: Rating distribution by Category
for cat, color in zip(['Food','Clothing','Electronics'], ['salmon','steelblue','seagreen']):
    sns.kdeplot(ax=axes[1, 1], data=df3[df3['Category'] == cat]['Rating'],
                label=cat, color=color, fill=True, alpha=0.3, linewidth=2)
axes[1, 1].set_title('Rating KDE by Category')
axes[1, 1].set_xlabel('Rating'); axes[1, 1].set_ylabel('Density')
axes[1, 1].legend(title='Category', fontsize=8)

# Panel 6 — Line Chart: Total Sales per Day (aggregated)
day_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
daily_sales = df3.groupby('Day')['Sales'].sum().reindex(day_order)
axes[1, 2].plot(day_order, daily_sales.values, marker='o', color='navy',
                linewidth=2.5, markersize=8)
axes[1, 2].fill_between(day_order, daily_sales.values, alpha=0.15, color='navy')
axes[1, 2].set_title('Total Sales by Day (Line Chart)')
axes[1, 2].set_xlabel('Day'); axes[1, 2].set_ylabel('Total Sales')
axes[1, 2].grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
